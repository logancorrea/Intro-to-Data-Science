### Task 1.1: Calculate the Sum of a Vector

Write a function that calculates and returns the sum of a vector that you pass into it.

```python
# Define function "vector_sum" that takes vector as input
def vector_sum(vector):

  # Initialize variable "sum" to store sum of vector
  sum = 0

  # Iterate over each element in vector and add current element to sum
  for i in vector:
    sum += i

  # Return sum rounded to two decimal places
  return round(sum,2)

print(vector_sum(temperature_vector))
print(vector_sum(precipitation_vector))
```

### Task 1.2: Calculate the Mean of a Vector

Write a function that calculates and returns the [arithmetic mean](https://en.wikipedia.org/wiki/Arithmetic_mean) of a vector that you pass into it. Use your above defined sum function to calculate sum of all the elements of a vector.

Pass the temperature and precipitaton vector into this function and print the result. Provide a written interpretation of your results (e.g., "The mean temperature for Snowbird is XXX degrees Fahrenheit. The mean precipitation is XXX inches.")

```python
# Define function "vector_mean" that takes vector as an input
def vector_mean(vector):

  # Calculate the sum of input vector using vector_sum function (defined in Task 1.1)
  element_sum = vector_sum(vector)

  # Calculate the mean of vector by dividing the sum of elements by the length of the vector
  mean = element_sum/len(vector)

  # Return the mean rounded to two decimal places
  return round(mean,2)

print(vector_mean(temperature_vector))
print(vector_mean(precipitation_vector))
```

### Task 1.3: Calculate the Median of a Vector
Write a function that calculates and returns the [median](https://en.wikipedia.org/wiki/Median) of a vector. Pass the temperature vector into this function and print the result. Make sure that your function works for both vectors with an even and odd number of elements. In the case of an even number of elements, use the mean of the two middle values. Provide a written interpretation of your results.

Hint: the [`sorted()`](https://docs.python.org/3/library/functions.html#sorted) function might be helpful for this.

```python
# Define function "vector_median" that takes vector as input
def vector_median(vector):

  # Sort vector from smallest to largest values
  vector_sorted = sorted(vector)

  # Find length of vector
  range = len(vector)

  # Iterate through elements in vector
  for i in vector_sorted:

    # Check if the length of the vector is odd
    if range % 2 == 1:

      # If odd, calculate middle index
      odd_mid = math.ceil(range/2)

      # Set median to middle element
      median = vector_sorted[odd_mid-1]

    # Check if the length of the vector is even
    elif range % 2 == 0:

      # If even, calculate the two middle indices
      even_lower = int(range/2)
      even_upper = even_lower + 1

      # Set median to the average of the middle indices
      median = (vector_sorted[even_lower-1] + vector_sorted[even_upper-1])/2

  # Return median rounded to two decimal places
  return round(median,2)

print(vector_median(temperature_vector))
print(vector_median(precipitation_vector))
```

### Task 1.4: Calculate the Variance of a Vector

Write a function that calculates and returns the [variance](https://en.wikipedia.org/wiki/Variance) of a vector. Pass the temperature and precipitation vectors into this function and print the result. Provide a written interpretation of your results. Is the variance high? Why, Why not?

The variance is the average of the squared deviations from the mean, i.e.,

$$ Var(X) = {\sigma}^{2} = \frac{1}{N} \sum_{i=1}^{N} {{(x_i - \mu)}^2} $$

where $\mu$ is the mean of the vector. Hint: use your mean function to calculate it.

```python
# Define function "vector_variance" that takes vector as input
def vector_variance(vector):
  # Intialize empty list "diff"
  diff = []

  # Caluclate the mean of the vector using the vector_mean function (defined in Task 1.2)
  mean = vector_mean(vector)

  # Iterate through through elements in vector
  for i in vector:

    # Calculate the the square root of the difference between the element and the mean at each element and append value to list "diff": (xi−μ)2
    diff.append((i-mean)**2)

    # Calculate the variance by dividing the sum of list "d"iff" by the length of list "diff"
    variance = sum(diff)/len(diff)

  # Return variance rounded to two decimal places.
  return round(variance,2)

print(vector_variance(temperature_vector))
print(vector_variance(precipitation_vector))
```

```python
print(np.var(temperature_vector))
print(np.var(precipitation_vector))
```

Here, the first array gives the size of these bins, the second defines the bands that are equally spaced between minimum and maximum value. That is, the first band from 21.8-28.75 has 20 entries, the second, from 28.75-35.7 has 16 entries, etc.


Calculate the histogram for precipitation and temperature when b=5. Once you have the calculated bins and values, plot the histogram for precipitaion and temperature using `hist` function in `matplotlib.pyplot` library. Provide a written interpretation of your results. Comment on the shape of each histogram; is it unimodal or bimodal? Is the distribution skewed (if so, in which direction)?

```python
import matplotlib.pyplot as plt

# Define function "histogram" that takes vector and bins as arguments
def histogram(vector,bins):

  # Initialize list "borders" with the smallest value in input vector
  borders = [min(vector)]

  # Calculate bin width based on the range of the vector divided by the number of bins
  bin_width = (max(vector)-min(vector))/(bins)

  # Initialize variable "value" as minimum value in vector
  value = min(vector)

  # Initialze empty list "counts"
  counts = []

  # Iterate through elements in the input vector
  for i in vector:

    # If value is less than the maximum value in vector:
    if value < max(vector):

      # Add the width of the bin to value and append to list "borders" rounded to two decimal places
      value += bin_width
      borders.append(round(value,2))

  # Iterate over bin borders to count occurrences in each bin
  for i in range(len(borders) - 1):

    # Initialize counter for current bin
    count = 0

    # Iterate over the vector to count occurrences within current bin
    for num in vector:

        # Check if the element is within the range bin range
        if borders[i] <= num < borders[i + 1]:

            # If true increment counter
            count += 1

    # Append count to list "counts" for each bin range
    counts.append(count)

  # Return list "counts" and list "borders"
  return [counts, borders]

# the call to your function
print(histogram(temperature_vector, bins = 5))
print(histogram(precipitation_vector,bins = 5))
```

### Part 3.1. Descriptive statistics

Describe what a Poisson random variable is. What is the parameter, $\lambda$? What is the min, max, mean, and variance of a Poisson random variable?

**Your description:**

A Poisson random variable is a discrete probability distrubution that represents the the number of events that occur within a fixed interval of time. Where λ is the expected number of events in a given interval, the probability of k events is:

$$ f(k; {\lambda}) = e^{-\lambda} {\lambda}^k/k!  $$

The minmum value of a poisson random variable is 0, there is no maximum value, and the mean and variance are equal to λ.

$$ {\lambda} = E(X) = Var(X) $$

### Part 3.3. Probability mass function

For the parameter $\lambda = 4$, plot the probability mass function (you may use scipy).

```python
from scipy.stats import poisson

# Set the parameter lambda
lmbda = 4

# Generate k values from 0 to 10
k_values = np.arange(0, 11)

# Calculate the PMF for each k value
pmf_values = poisson.pmf(k_values, lmbda)

# Plot the PMF
plt.bar(k_values, pmf_values)
plt.title('Poisson PMF for $\lambda=4$')
plt.xlabel('Number of Events (k)')
plt.ylabel('Probability')

plt.show()
```

### Part 3.4. Poission sampling

Write python code that takes 1500 samples from the Poisson distribution with parameter $\lambda = 4$. Make a histogram of the samples and compute the sample mean and variance. How does the histogram compare to the probability mass function?

```python
# Set parameters lambda and k
lmbda = 4
k = 1500

# Generate random values from Poisson distribution
poisson_dist = poisson.rvs(mu=lmbda, size=k)

# Calculate distribution mean and variance
dist_mean = np.mean(poisson_dist)
dist_variance = np.var(poisson_dist)

# Generate histogram of Poisson distibution samples
plt.hist(poisson_dist, bins = 10)
plt.title("Histogram of Poisson Distibution")
plt.ylabel("Samples")
plt.xlabel("Sample Occurrence")
plt.show

print("Poisson Distibution Mean: " + str(dist_mean))
print("Poisson Distribution Variance: " + str(dist_variance))
```

**Your description:** The Poisson distribution looks similar to the probability pass function with the sample average around 4 occurences.
